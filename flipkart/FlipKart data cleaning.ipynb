```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('phone_dataset.csv')
```

```python
df.head()
```

# Price, Ratings and Review colums are object type, need to clean data so it can be converted to Integer data type

```python
df.info()
```

## Droped caloums which are not needed

```python
df.drop(columns=['Variant', 'Display','Camera','Offers'], inplace = True)
```

## Removing coma & ₹ symbol using string replace function

```python
df['Price'].str.replace(',', '').str.replace('₹', '')
```

## Same way removed coma from all the columns which we want to convert from String to Integer data type

```python
df['Price'] = df['Price'].str.replace(',', '').str.replace('₹', '')
df['5 star rating'] = df['5 star rating'].str.replace(',', '')
df['4 star rating'] = df['4 star rating'].str.replace(',', '')
df['3 star rating'] = df['3 star rating'].str.replace(',', '')
df['2 star rating'] = df['2 star rating'].str.replace(',', '')
df['1 star rating'] = df['1 star rating'].str.replace(',', '')
df['Total no of reviews'] = df['Total no of reviews'].str.replace(',', '').str.replace('Reviews', '')
df['Total no of rating'] = df['Total no of rating'].str.replace(',', '').str.replace('(', '').str.replace(')', '')
```

## Data is cleaned now

```python
df.head(2)
```

## Converting String data type to integer

```python
df['Price'] = df['Price'].astype(int)
df['5 star rating'] = df['5 star rating'].astype(int)
df['4 star rating'] = df['4 star rating'].astype(int)
df['3 star rating'] = df['3 star rating'].astype(int)
df['2 star rating'] = df['2 star rating'].astype(int)
df['1 star rating'] = df['1 star rating'].astype(int)
df['Total no of reviews'] = df['Total no of reviews'].astype(int)
df['Total no of rating'] = df['Total no of rating'].astype(int)
```

```python
df.info()
```

```python
df.columns
```

## Need to remove all duplicate values for accurate insight gathering

```python
df.drop_duplicates(subset=['Name'],inplace=True)
```

```python
df.info()
```

### resetting the index

```python
df.reset_index(inplace=True)
```

```python
df.tail()
```
